In this notebook, we'll be learning **K-Means Clustering**, a popular **unsupervised learning method** for **clustering similar data based on the Euclidean distance between samples.**


````python
# import packages
import copy
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
# plotting set-up
plt.rc('font', family = 'serif', size = 13, weight = "bold")
# My Set-Up
plt.rc('xtick', labelsize = 10)
plt.rc('ytick', labelsize = 10)
# Sam's Set-Up
# plt.rc('xtick', labelsize = 10, direction = 'in', top = 'on')
# plt.rc('xtick.major', size = 5, width = 2)
# plt.rc('ytick', labelsize = 10, direction = 'in', right = 'on')
# plt.rc('ytick.major', size = 5, width = 2)
plt.rc('axes', linewidth = 2)
plt.rc('lines', linewidth = 3)
plt.rc('figure', figsize = (6, 6))
plt.rc('legend', fontsize = 13)

cm = plt.cm.get_cmap('tab20')
````

We will start with **generating some blob cluster dataset.**

````python
from sklearn.datasets import make_blobs

blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])
blob_std = np.array([0.2, 0.3, 0.11, 0.12, 0.1])

X, y = make_blobs(n_samples = 3000, centers = blob_centers,
                  cluster_std = blob_std, random_state = 89)
````

````python
fig = plt.figure(figsize = (8, 4))

plt.scatter(X[:, 0], X[:, 1], s = 1)

plt.xlabel("$x_1$", fontsize = 14)
plt.ylabel("$x_2$", fontsize = 14)

plt.title("Sample cluster dataset", weight = "bold")

plt.show()
````

````python
from sklearn.cluster import KMeans
````

````python
help(KMeans)
````

````python
k = 5 # number of clusters
kmeans = KMeans(n_clusters = k, random_state = 42)
````

````python
y_pred = kmeans.fit_predict(X)
````

````python
kmeans.cluster_centers_
````

### Hard Clustering

We first visualize the **"hard clustering" result**, where each sample either belongs to a cluster **completely or not.** 

````python
import seaborn as sns

fig = plt.figure(figsize = (8, 4))
sns.scatterplot(x = X[:, 0], y = X[:, 1], hue = y_pred,\
                palette = sns.hls_palette(k), legend = 'full')

plt.title("Hard Cluster Result with K-Means", weight = "bold", fontsize = 12)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
````

### Decision Boundary

````python
fig = plt.figure(figsize = (8, 4))

mins = X.min(axis = 0) - 0.1
maxs = X.max(axis = 0) + 0.1
xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], 1000),
                     np.linspace(mins[1], maxs[1], 1000))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(Z, extent = (mins[0], maxs[0], mins[1], maxs[1]),
            cmap = "Accent")
# plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
#             linewidths=1, colors='k')

sns.scatterplot(x = X[:, 0], y = X[:, 1], hue = y_pred,\
                palette = sns.hls_palette(k), legend = 'full')

plt.title("K-Means with Decision Boundaries", weight = "bold", fontsize = 12)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
````

### Soft Clustering and ```transform```

Instead of hard clustering, we can also do **soft clustering**,- using a **probability or likelihood** of that data point to be in those clusters is assigned. 

For example, We can use the **distance to each cluster centroid** as a matrix for calculating the probability.

Moreover, ```kmeans.transform()``` calculates the **distance to each cluster centroid**

````python
kmeans.transform(X[0].reshape(1, -1))
````

````python
def soft_clustering(dist):
    """
    return the soft clustered probability by:
    
    calculating the inversed distance divided by the sum.
    
    """
    inversed_distance = np.array([1/i for i in dist])

    return inversed_distance / inversed_distance.sum()

soft_clustering(kmeans.transform(X[0].reshape(1, -1)))
````

### Measuring K-Means performance with ```inertia```

Kmeans measurement matrix: inertia metric -  **measure the distance between each instance and its centroid.**

In practice, we'd want to minimize this value

````python
kmeans.inertia_
````

K-Means can be optimized by **running the K-Means algorithm multiple times with different random initializations**, and select the solution that **minimizes the inertia.**

- Default value for n_init is 10 times, and can be increased by the ```n_init```

### Elbow Plot and Inertia

Inertia can also be used as a matrix to determine whether ```k``` clusters are enough.

````python
kmeans_per_k = [KMeans(n_clusters = k, random_state = 42).fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]
````

````python
plt.figure(figsize=(6, 4), dpi = 80)

plt.plot(range(1, 10), inertias, "bo-")

plt.xlabel("$k$", weight = "bold")
plt.ylabel("Inertia", weight = "bold")
plt.annotate('Elbow',
             xy=(4, inertias[3]),
             xytext=(0.55, 0.55),
             textcoords='figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )

plt.title("Elbow Plot for K Determination", weight = "bold", fontsize = 12)

plt.show()
````
